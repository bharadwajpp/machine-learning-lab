```python
def calculate_cost(x, y, w, b):

    m = len(x)
    y_pred = w * x + b
    cost = (1 / (2 * m)) * np.sum((y_pred - y) ** 2)
    return cost
```

```python
cost = calculate_cost(x, y, w=100, b=100)
print(f"Cost for w=8, b=0: {cost:.2f}")
```

```python
 def calculate_gradient(x, y, w, b):

    m = len(x)
    y_pred = w * x + b
    dw = (1 / m) * np.sum((y_pred - y) * x)  # derivative w.r.t w
    db = (1 / m) * np.sum(y_pred - y)        # derivative w.r.t b
    return dw, db
```

```python
dw, db = calculate_gradient(x, y, w, b)
print(dw)
print(db)
```

```python
def gradient_descent(x, y, w_init, b_init, learning_rate, iterations):
    w = w_init
    b = b_init
    cost_history = []

    for i in range(iterations):
        dw, db = calculate_gradient(x, y, w, b)
        w -= learning_rate * dw
        b -= learning_rate * db
        cost = calculate_cost(x, y, w, b)
        cost_history.append(cost)

    return w, b, cost_history
```

```python
# Parameters for gradient descent
w_init = 5
b_init = 30
learning_rate = 0.01
iterations = 100

# Run gradient descent
w_opt, b_opt, cost_history = gradient_descent(x, y, w_init, b_init, learning_rate, iterations)

# Print optimized values
print(f"Optimized weight (w): {w_opt:.4f}")
print(f"Optimized bias (b): {b_opt:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([45, 50, 60, 65, 80, 95])

def calculate_cost(x, y, w, b):
    m = len(x)
    y_pred = w * x + b
    cost = (1 / (2 * m)) * np.sum((y_pred - y) ** 2)
    return cost


w_vals = np.linspace(50, 200, 50)
b_vals = np.linspace(-200, 200, 50)


J_vals = np.zeros((len(w_vals), len(b_vals)))


for i in range(len(w_vals)):
    for j in range(len(b_vals)):
        J_vals[i, j] = calculate_cost(x, y, w_vals[i], b_vals[j])


W, B = np.meshgrid(w_vals, b_vals)


fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(W, B, J_vals.T, cmap="viridis")
ax.set_title("Cost Function Surface")
ax.set_xlabel("Weight (w)")
ax.set_ylabel("Bias (b)")
ax.set_zlabel("Cost")
plt.show()
```
